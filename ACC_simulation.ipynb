```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
dt = 0.1
time = np.arange(0, 40, dt)

# Initial positions
x_lead = 50
x_ego = 0

# Initial speeds
v_lead = 20
v_ego = 22

T_gap = 2

# Data storage
lead_positions = []
ego_positions = []
distances = []
safe_distances = []
ego_speeds = []
lead_speeds = []

for t in time:

    # Lead vehicle slows after 10 seconds
    if t > 10:
        v_lead = 15

    # Update lead vehicle position
    x_lead += v_lead * dt

    # Calculate distance
    distance = x_lead - x_ego

    # Safe distance
    d_safe = v_ego * T_gap

    # ACC logic
    if distance < d_safe:
        v_ego -= 1 * dt
    elif distance > d_safe + 5:
        v_ego += 1 * dt

    # Update ego vehicle position
    x_ego += v_ego * dt

    # Collision warning
    if distance < 5:
        print("Collision Warning!")

    # Store data
    lead_positions.append(x_lead)
    ego_positions.append(x_ego)
    distances.append(distance)
    safe_distances.append(d_safe)
    ego_speeds.append(v_ego)
    lead_speeds.append(v_lead)

# -------- Graph 1 Distance vs Safe Distance --------
plt.figure()
plt.plot(time, distances, label="Actual Distance")
plt.plot(time, safe_distances, label="Safe Distance")
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.title("ACC Distance Control")
plt.legend()
plt.show()

# -------- Graph 2 Speed Comparison --------
plt.figure()
plt.plot(time, ego_speeds, label="Ego Vehicle Speed")
plt.plot(time, lead_speeds, label="Lead Vehicle Speed")
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.title("Vehicle Speed Comparison")
plt.legend()
plt.show()
```
